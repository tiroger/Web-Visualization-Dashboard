```python
# Importing ddependencies and packages for analysis
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import geopandas as gpd
import bokeh

import datetime
```

```python
collisions_df = pd.read_csv('Resources/Cleaned_NYPD_Motor_Vehicle_Collisions.csv', low_memory=False)
```

```python
collisions_df.head()
```

```python
# Converting date to datetime object and extracting year in new column
collisions_df['DATE'] = pd.to_datetime(collisions_df['DATE'])
collisions_df['YEAR'] = collisions_df['DATE'].dt.year
collisions_df.head()
```

## Analysis

```python
# Grouping data by year
collisions_by_year = collisions_df.groupby('YEAR').sum()
collisions_by_year.head()
```

```python
# Grouping data by zip code
collisions_by_zip = collisions_df.groupby('ZIP CODE').sum()
collisions_by_zip.head()
```

```python
collisions_df['BOROUGH'].unique()
```

```python
# Grouping data by borough
collisions_by_boro = collisions_df.groupby('BOROUGH').sum()
collisions_by_boro.head(10)
```

## We define the following function
def haversine(lat1, lon1, lat2, lon2):
    MILES = 3959
    lat1, lon1, lat2, lon2 = map(np.deg2rad, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1 
    dlon = lon2 - lon1 
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a)) 
    total_miles = MILES * c
    return total_miles

```python
# Calculating bike path lenght
# Function to calculate length based on coordinates
from bike_lane_length_calculator import haversine 
lane_lengths = []

for i in range(len(bike_lanes_df["lane_start"])):
    lat1 = float(bike_lanes_df.iloc[i]['lat1'])
    lon1 = float(bike_lanes_df.iloc[i]['lon1'])
    lat2 = float(bike_lanes_df.iloc[i]['lat2'])
    lon2 = float(bike_lanes_df.iloc[i]['lon2'])
    
    total_miles = haversine(lat1, lon1, lat2, lon2)
    lane_lengths.append(total_miles)
```
